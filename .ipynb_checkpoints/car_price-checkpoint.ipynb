````python
# чтобы не создавать каждый раз функцию для отображения числовых данных, напишу следующие функции
def histogram(column_name):
    sns.histplot(data=df[column_name], bins=20, kde=True)
    plt.title(column_name+' distribution')
    plt.show()
````

````python
# аналогично сделаю для категориальных данных
def count_plot(column):
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x=column)
    plt.title(f"Count Plot - {column}")
    plt.xlabel(column)
    plt.ylabel("Count")
    plt.xticks(rotation=90)
    plt.show()
````

````python
from statsmodels.formula.api import ols
model = ols("price ~ enginesize + wheelbase", data=df1)
res = model.fit()
print(res.summary())
````

Так, мы получили следующее уравнение множественной линейной регрессии:

```
y = -3.337e+04 + 101.2020  * x1 + 334.7283 * х2
```


1) При изменении размере двигателя на 1 см3 цена на автомобиль в среднем изменяется на 101.2020 долл. США в ту же сторону при неизменном значении колесной базы автомобиля

2) При изменении колесной базы автомобиля на 1 см цена на автомобиль в среднем изменяется на  334.7283 долл. США  в ту же сторону при неизменном размере двигателя.
* wheelbase - колесная база автомобиля (Numerical)
* enginesize - Размер двигателя автомобиля(Numerical)

### 3. Рассчитаю стандартизованные коэф-ты регрессии

````python
corr = df1.corr()
corr
````

````python
#
B1 = round((corr.iloc[1][0] - corr.iloc[2][0] * corr.iloc[1][2]) / (1 - corr.iloc[1][2] * corr.iloc[1][2]), 3)
B2 = round((corr.iloc[2][0] - corr.iloc[1][0] * corr.iloc[1][2]) / (1 - corr.iloc[1][2] * corr.iloc[1][2]), 3)
print(f"B1: {B1},\nB2: {B2}")
````

=> **<font color='limegreen'>ty = 0.126 * tx1 + 0.666 * tx2</font>**

1) С увеличением размере двигателя на на 1 σ и неизменном размере колесной базы цена автомобиля в ср. вырастет на 0,126 σ

2) С увеличением размера колесной базы на 1 σ и неизменном размере двигателя цена автомобиля  в ср. вырастет на 0,666 σ, что оказывает наибольшее влияние на признак-результат

### 4. Рассчитаем коэф-ты эластичности

````python
y_mean = df1['price'].mean()
x1_mean = df1['enginesize'].mean()
x2_mean = df1['wheelbase'].mean()
````

````python
Eyx1 = 0.1292 * x1_mean / y_mean
Eyx2 = 0.0033 * x2_mean / y_mean
print(f'Эyx1: {Eyx1},\nЭyx2: {Eyx2}')
````

Эух1 показывает, что цена на автомобиль увеличится на 0.0013 %  при изменении размера двигателя на 1%.

Эух2 показывает, что цена на автомобиль увеличится на 0.00003%  при изменении размера колесной базы на 1%.


### 5. Оценим значимость параметров и уравнения множественной регрессии.

* tтабл(a=0,05; df=195-2-1) = 1.96
    
    |ta|< tкрит => параметр а статистически не значим c вероятностью 0.95(однако это интерсепт, поэтому незначимость параметра допустима)

    |tb1| > tкрит => параметр b1 статистически значим  с вероятностью 0.95
    
    |tb2|> tкрит  => параметр b2 статистически значим с вероятностью 0.95

* Fкрит =  174.3
* Fтабл = 3.04
=> Значение F критерия намного выше табличного, поэтому уравнение статистическо значимо с вер. 0.95

### 6.  Оценим значимость включения переменных в уравнение регрессии¶

````python
R2 = 0.645
F1 = (R2 - corr.iloc[1][0])/(1-R2) * 192
F2 = (R2 - corr.iloc[2][0])/(1-R2) * 192
F1, F2
````

Дополнительное включение фактора x1 в модель статистически оправданно. Коэффициент b1 статистически значим с вер. 0.95

Дополнительное включение фактора x2 в модель статистически оправданно. Коэффициент  b2 статистически значим с вер. 0.95

* R2yx1x2 = 0.645, то есть 64,5% вариации фактора-результата(цены автомобиля) объясняется варицией размера двигателя и колесной базы

### 11. Исследуем остатки на гетероскедастичность для множественной регрессии без фиктивных переменных и с фиктивными переменными.

````python
# без dummy variables
df2['price'] = res.resid

sns.pairplot(df2, y_vars = 'price', x_vars = ['enginesize', 'wheelbase', 'citympg'], size = 5, diag_kind=None)
````

````python
y = df2['price']
x = df2[['enginesize', 'wheelbase', 'citympg']]
x = sm.add_constant(x)
model = sm.OLS(y, x).fit()
````

````python
# Тест Уайта с dummy variables
white_test = het_white(model.resid , model. model.exog )
labels = ['Test Statistic', 'Test Statistic p-value', 'F-Statistic', 'F-Test p-value']
print(dict(zip(labels, white_test)))
````

Таким образом, т.к. p-value < 0.05 =>  отвергается нулевая о наличии гомоскедастичности => остатки гетероскедастичны

````python
# аналогично без фиктивных переменных
y = df1['price']
x = df1[['enginesize', 'wheelbase']]
x = sm.add_constant(x)
model = sm.OLS(y, x).fit()
````

````python
white_test = het_white(model.resid , model. model.exog )
labels = ['Test Statistic', 'Test Statistic p-value', 'F-Statistic', 'F-Test p-value']
print(dict(zip(labels, white_test)))
````

Таким образом, т.к. p-value < 0.05 =>  отвергается нулевая о наличии гомоскедастичности => остатки гетероскедастичны
