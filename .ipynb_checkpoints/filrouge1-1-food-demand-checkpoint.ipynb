```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_squared_log_error, mean_absolute_error, confusion_matrix, make_scorer
from sklearn.model_selection import StratifiedKFold, cross_val_score
import xgboost as xgb

```

```python
#Preprocessing for model building
```

```python
# Select the columns for feature encoding and scaling
categorical_cols = train_combined.select_dtypes(include=['object', 'category']).columns.tolist()
numerical_cols = train_combined.select_dtypes(include=['int64', 'float64']).columns.tolist()
numerical_cols.remove('num_orders')  # Exclude the target variable from scaling
```

```python
# Define the preprocessor
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols),
        ('cat', OneHotEncoder(), categorical_cols)
    ]
)
```

```python
#Model choosing and building
#Models chose were Random Forest Regressor and Multi Layer Perceptron Regressor (neural network) and xgboost regressor
```

```python
# Define the neural network model
nn_model = MLPRegressor(hidden_layer_sizes=(100,50), max_iter=600, random_state=101)
```

```python
# Define the random forest model
rf_model = RandomForestRegressor(n_estimators=100, random_state=101)
```

```python
# Define the xgboost model
xgb_model = xgb.XGBRegressor(n_estimators=100, random_state=101)
```

```python
# Create the pipeline for the neural network model
pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                           ('model', nn_model)])
```

```python
# Create the pipeline for random forest
pipeline_rf = Pipeline(steps=[('preprocessor', preprocessor),
                              ('model', rf_model)])
```

```python
# Create the pipeline for xgboost
pipeline_xgb = Pipeline(steps=[('preprocessor', preprocessor),
                                 ('model', xgb_model)])
                                

                                
```

```python
# Separate features and target variable
X = train_combined.drop('num_orders', axis=1)
y = train_combined['num_orders']
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
# Train the neural network model
pipeline.fit(X_train, y_train)
```

```python
# Train the random forest model
pipeline_rf.fit(X_train, y_train)
```

```python
# Train the xgboost model
pipeline_xgb.fit(X_train, y_train)
```

```python
# Predictions for neural network ,random forest and xgboost
y_pred_nn = pipeline.predict(X_test)
y_pred_rf = pipeline_rf.predict(X_test)
y_pred_xgb = pipeline_xgb.predict(X_test)
```

```python
# Plotting model comparison
```

```python
# 3 way comparison plot
# Plot for neural network

plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
plt.scatter(y_test, y_pred_nn, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.title(f'Neural Network Model\nR-squared: {r2_nn:.2f} | RMSE: {rmse_nn:.2f}')
plt.xlabel('Actual')
plt.ylabel('Predicted')

# Plot for random forest
plt.subplot(2, 2, 2)
plt.scatter(y_test, y_pred_rf, alpha=0.5, color='green')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.title(f'Random Forest Model\nR-squared: {r2_rf:.2f} | RMSE: {rmse_rf:.2f}')
plt.xlabel('Actual')
plt.ylabel('Predicted')

# Plot for xgboost
plt.subplot(2, 2, 3)
plt.scatter(y_test, y_pred_xgb, alpha=0.5, color='red')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.title(f'Xgboost Model\nR-squared: {r2_xgb:.2f} | RMSE: {rmse_xgb:.2f}')
plt.xlabel('Actual')
plt.ylabel('Predicted')



plt.tight_layout()
plt.show()
```

```python
#Interpretation
# The above plot can be explained in the following way: 
# The closer the points are to the diagonal line, the better the model's predictions are.
# The scatter plots show that the random forest model has a better fit than the neural network model. 
# The random forest model has a higher R-squared value and a lower RMSE value than the neural network model. 
# The random forest model also has a higher correlation (r squared) between the actual and predicted values than the neural network model.
```

```python
# Cross validation for random forest
scores_rf = cross_val_score(pipeline_rf, X, y, cv=5, scoring='neg_mean_squared_error')
rmse_scores_rf = np.sqrt(-scores_rf)
print('Random Forest RMSE scores: ', rmse_scores_rf)
print('Random Forest RMSE mean: ', rmse_scores_rf.mean())
print('Random Forest RMSE standard deviation: ', rmse_scores_rf.std())
```

```python
# Cross validation for neural network
scores_nn = cross_val_score(pipeline, X, y, cv=5, scoring='neg_mean_squared_error')
rmse_scores_nn = np.sqrt(-scores_nn)
print('Neural Network RMSE scores: ', rmse_scores_nn)
print('Neural Network RMSE mean: ', rmse_scores_nn.mean())
print('Neural Network RMSE standard deviation: ', rmse_scores_nn.std())
```

```python
# Cross validation for XGBoost 
scores_xgb = cross_val_score(pipeline_xgb, X, y, cv=5, scoring='neg_mean_squared_error')
rmse_scores_xgb = np.sqrt(-scores_xgb)
print('XGBoost RMSE scores: ', rmse_scores_xgb)
print('XGBoost RMSE mean: ', rmse_scores_xgb.mean())
print('XGBoost RMSE standard deviation: ', rmse_scores_xgb.std())
```

```python
# RMSLE scores for all models (metric used for evaluation in kaggle competition)
rmsle_scores_rf = 100*np.sqrt(mean_squared_log_error(y_test, y_pred_rf))
rmsle_scores_nn = 100*np.sqrt(mean_squared_log_error(y_test, y_pred_nn))
rmsle_scores_xgb = 100*np.sqrt(mean_squared_log_error(y_test, y_pred_xgb))

print('Random Forest RMSLE score: ', rmsle_scores_rf)
print('Neural Network RMSLE score: ', rmsle_scores_nn)
print('XGBoost RMSLE score: ', rmsle_scores_xgb)
```

```python
# R2 scores for all models
r2_scores_rf = r2_score(y_test, y_pred_rf)
r2_scores_nn = r2_score(y_test, y_pred_nn)
r2_scores_xgb = r2_score(y_test, y_pred_xgb)

print('Random Forest R2 score: ', r2_scores_rf)
print('Neural Network R2 score: ', r2_scores_nn)
print('XGBoost R2 score: ', r2_scores_xgb)
```

```python
# Based on the RMSLE scores and cross validation scores, the random forest model is the best model for predicting the number of orders.
# R2 explains the variance in the target variable, and the random forest model has the highest R2 score, 
# indicating that it explains the variance in the target variable the best. (85%)
```

```python
# Making plot of comparison of scores
```

```python
# Data from the table
models = ['Random Forest', 'Neural Network', 'XGBOOST']
rmse_mean = [202.94, 195.21, 207.27]
rmse_std = [28.38, 28.50, 42.38]
r2 = [.85, .82, .84]
rmsle = [48.89, 74.79, 73.48]

# Colors and winner's highlight
colors = ['grey' if i != min(rmse_mean) else 'green' for i in rmse_mean]
std_colors = ['grey' if i != min(rmse_std) else 'green' for i in rmse_std]
r2_colors = ['grey' if i != max(r2) else 'blue' for i in r2]
rmsle_colors = ['grey' if i != min(rmsle) else 'blue' for i in rmsle]

# Function to add value labels
def add_labels(x, y, ax):
    for i in range(len(x)):
        ax.text(i, y[i], f"{y[i]:.2f}", ha = 'center', va = 'bottom')

# Plotting the RMSE Mean
plt.figure(figsize=(14, 8))

# Subplot for RMSE Mean
ax1 = plt.subplot(2, 2, 1)
ax1.bar(models, rmse_mean, color=colors)
ax1.set_title('RMSE Mean (Cross Validation)', fontsize=14, color='black')
ax1.set_ylabel('RMSE Mean')
add_labels(models, rmse_mean, ax1)

# Subplot for RMSE Standard Deviation
ax2 = plt.subplot(2, 2, 2)
ax2.bar(models, rmse_std, color=std_colors)
ax2.set_title('RMSE Standard Deviation (Cross Validation)', fontsize=14, color='black')
ax2.set_ylabel('RMSE Standard Deviation')
add_labels(models, rmse_std, ax2)

# Subplot for R2 Score
ax3 = plt.subplot(2, 2, 3)
ax3.bar(models, r2, color=r2_colors)
ax3.set_title('R2 Score (Without Cross Validation)', fontsize=14, color='black')
ax3.set_ylabel('R2 Score')
add_labels(models, r2, ax3)

# Subplot for RMSLE
ax4 = plt.subplot(2, 2, 4)
ax4.bar(models, rmsle, color=rmsle_colors)
ax4.set_title('RMSLE (Without Cross Validation)', fontsize=14, color='black')
ax4.set_ylabel('RMSLE')
add_labels(models, rmsle, ax4)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# The random forest model has the lowest RMSE mean and standard deviation, the highest R2 score, and the lowest RMSLE score.
# Therefore, the random forest model is the best model for predicting the number of orders.
# Green metrics were calculated using cross validation, and blue metrics were calculated without cross validation.
```
